```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
sns.heatmap(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X - scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
# PCA N Feature Space ---> N PC Space
# How ,any PC to choose? N = 2
```

```python
num_componenets = 2
```

```python
np.argsort([2, 1, 3])
```

```python
np.argsort(eigen_values)
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_componenets]
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:, sorted_key]
```

```python
# Original Data --> project --> eigen_vectors
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(8, 6), dpi=150)
plt.scatter(principal_components[:, 0], principal_components[:, 1])
```
