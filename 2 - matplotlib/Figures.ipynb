```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
```

```python
b = a**4
```

```python
fig = plt.figure()

#Large axes

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b);
axes.set_xlim(0, 8)
axes.set_ylim(0, 8000)
axes.set_xlabel('A')
axes.set_ylabel('B')
axes.set_title('Power of 4')

#Small axes

axes2 = fig.add_axes([1, 1, 0.25, 0.25])
axes2.plot(a, b);
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

```python
fig = plt.figure(figsize=(2,2), dpi=200)

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight')
```
