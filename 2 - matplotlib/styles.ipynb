```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2,label='X vs X^2')

ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#possible linestype options '--', '-', '-.', ':', 'steps'

ax.plot(x,x,color='#ff36bc', label='x vs x', lw=5, ls=':')

ax.plot(x, x+1, color='#fcfc47', label='x vs x+1', linewidth=2, linestyle='-.')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x, x+1, color='#fcfc47', label='x vs x+1', linewidth=4)
lines[0].set_dashes([1,1,1,2,3,5])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color='red', label='x vs x+1', linewidth=2, marker='o', ls='--', ms=8, markerfacecolor='white', markeredgewidth=4, markeredgecolor='orange')

ax.plot(x, x, color='blue', label='x vs x', linewidth=2, marker='+', ls='--', markersize=14)
```
