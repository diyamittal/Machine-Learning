```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df
```

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education',split=True, bw=0.1)
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.35, 0.5))
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df, hue='gender')
```
