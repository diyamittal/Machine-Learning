```python
sns.pairplot(df)
```

```python
g = sns.PairGrid(df, hue='gender')

g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot) #scatterplot doesn't make sense
g = g.map_lower(sns.kdeplot, fill=True)
g = g.add_legend()
```
